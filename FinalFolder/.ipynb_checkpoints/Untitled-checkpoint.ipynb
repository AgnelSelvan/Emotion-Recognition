```python
import cv2
import numpy as np

name = input("Enter your name")
print(name)
path = './images/'+name+"/"
isPathExists = os.path.exists(path)
print(isPathExists)
if(isPathExists):
    pass
else:
    os.makedirs(path)

# Load HAAR face classifier
face_classifier = cv2.CascadeClassifier('../HaarCascade/haarcascade_frontalface_default.xml')

def face_extractor(img):
    # Function detects faces and returns the cropped face
    # If no face detected, it returns the input image
    
    #gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(img, 1.3, 5)
    
    if faces == ():
        return None
    
    # Crop all faces found
    for (x,y,w,h) in faces:
        x=x-10
        y=y-10
        cropped_face = img[y:y+h+50, x:x+w+50]

    return cropped_face

# Initialize Webcam
cap = cv2.VideoCapture(1)
count = 0

# Collect 100 samples of your face from webcam input
while True:

    ret, frame = cap.read()
    if face_extractor(frame) is not None:
        count += 1
        face = cv2.resize(face_extractor(frame), (400, 400))
        #face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

        # Save file in specified directory with unique name
        file_name_path = path + str(count) + '.jpg'
        cv2.imwrite(file_name_path, face)

        # Put count on images and display live count
        cv2.putText(face, str(count), (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
        cv2.imshow('Face Cropper', face)
        
    else:
        cv2.putText(frame, "Face Not Found", (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,255), 2)
        cv2.imshow('Face Cropper', frame)
        print("Face not found")
        pass

    if cv2.waitKey(1) == 13 or count == 100: #13 is the Enter Key
        break
        
cap.release()
cv2.destroyAllWindows()      
print("Collecting Samples Complete")
```

```python
def get_face_encoding():
    current_ids = 0
    x_train = []
    y_labels = []
    label_ids = {}
    
    for root, dirs, files in os.walk('./images'):
        print(files)
        for file in files:
            if file.endswith("png") or file.endswith("jpg"):
                path = os.path.join(root, file)
                print(path)
                try:
                    face = fr.load_image_file(path)
#                     print(face)
                    encoding = fr.face_encodings(face)[0]
                    dirname = root.split(os.path.sep)[-1]
                    print(dirname)

                    if dirname in label_ids.values():
                        pass
                    else:
                        current_ids += 1

                    label_ids[current_ids] = dirname
                    x_train.append(encoding)
                    y_labels.append(current_ids)
                except:
                    pass
                
    return x_train, y_labels, label_ids

x_train, y_labels, label_ids = get_face_encoding()
print(x_train, y_labels)
print(label_ids)
```

```python
import numpy as np
import pickle
recognizer = cv2.face.LBPHFaceRecognizer_create()

with open("./trainedmodel/label.pickle", "wb") as f:
    pickle.dump(label_ids, f)

recognizer.train(x_train, np.array(y_labels))
recognizer.save("./trainedmodel/trainner.yml")
```

```python
import os
import numpy as np
import face_recognition as fr
import cv2

path = './image/'
face_names = []
def get_all_encoded_faces():
    current_ids = 0
    encodings = {}
    
    for root, dirs, files in os.walk('./image'):
        print(files)
        for file in files:
            if file.endswith("png") or file.endswith("jpg"):
                path = os.path.join(root, file)
                print(path)
                try:
                    face = fr.load_image_file(path)
                    encoding = fr.face_encodings(face)[0]
                    filename = os.path.splitext(file)[0]
                    
                    encodings[filename] = encoding

                except:
                    pass
    
    print(encodings)
    return encodings

def face_extractor(img):
    
    faces = fr.face_locations(img)
    
    if faces == ():
        return None
    
    for (top, right, bottom, left) in faces:
        top = top - 50
        left =left - 50
        cropped_face = img[top:top+bottom-10, left:left+right - 10]

    return cropped_face

import time
name = "Unknown"
encoded_faces = {}
encoded_faces = get_all_encoded_faces()
faces_encoded = list(encoded_faces.values())
known_face_names = list(encoded_faces.keys())

cap = cv2.VideoCapture(1)

while True:
    ret, frame = cap.read()
    
    face_locations = fr.face_locations(frame)
    unknown_face_encodings = fr.face_encodings(frame, face_locations)
    
    face_names.append(name)
    
    if encoded_faces != {}:
    
        face_names = []
        for face_encoding in unknown_face_encodings:
            
            matches = fr.compare_faces(faces_encoded, face_encoding)
            
            face_distances = fr.face_distance(faces_encoded, face_encoding)
            best_match_index = np.argmin(face_distances)
            if matches[best_match_index]:
                name = known_face_names[best_match_index]

            face_names.append(name)
            print(face_names)
            
            if(name == "Unknown"):
                
                face = face_extractor(frame)
                face = cv2.resize(face, (500, 500))
                cv2.imshow("Unknown Image", face)
            else:
                for (top, right, bottom, left), name in zip(face_locations, face_names):
                    # Draw a box around the face
                    cv2.rectangle(frame, (left-20, top-20), (right+20, bottom+20), (255, 0, 0), 2)

                    # Draw a label with a name below the face
                    cv2.rectangle(frame, (left-20, bottom -15), (right+20, bottom+20), (255, 0, 0), cv2.FILLED)
                    font = cv2.FONT_HERSHEY_DUPLEX
                    cv2.putText(frame, name, (left -20, bottom + 15), font, 1.0, (255, 255, 255), 2)
                cv2.imshow("Img", frame)
    
    else:
        if name == "Unknown":
            print("Unknown Face Detected")
            face = face_extractor(frame)
            filename = input("Enter the unknown photo name: ")
            file_name_path = path + filename + '.jpg'
            img = cv2.imwrite(file_name_path, face)
            cv2.imshow("Image", img)
            encoded_faces = get_all_encoded_faces()
            faces_encoded = list(encoded_faces.values())
            known_face_names = list(encoded_faces.keys())
            break
            
# #             face = face_extractor(frame)
#             for (top, right, bottom, left), name in zip(face_locations, face_names):
#                 # Draw a box around the face
#                 cv2.rectangle(frame, (left-20, top-20), (right+20, bottom+20), (255, 0, 0), 2)
#                 cv2.rectangle(frame, (left-20, bottom -15), (right+20, bottom+20), (255, 0, 0), cv2.FILLED)
#                 font = cv2.FONT_HERSHEY_DUPLEX
#                 cv2.putText(frame, name, (left -20, bottom + 15), font, 1.0, (255, 255, 255), 2)
            
#             cv2.imshow("Unknown Image", frame)
            
            
#             filename = input("Enter the unknown photo name: ")
#             file_name_path = path + filename + '.jpg'
#             cv2.imwrite(file_name_path, face)
            
                
    if cv2.waitKey(1) == 13:
        break
        
cap.release()
cv2.destroyAllWindows()
```
